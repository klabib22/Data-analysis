```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
stars = pd.read_csv("stars.csv")
stars
```

```python
Tvec = np.array(stars[stars.columns[0]])
print(Tvec)
```

```python
Lvec = np.array(stars[stars.columns[1]])
print(Lvec)
```

```python
Rvec = np.array(stars[stars.columns[2]])
print(Rvec)
```

```python
Mvec = np.array(stars[stars.columns[3]])
print(Mvec)
```

```python
stardata = list(zip(Tvec,Lvec,Rvec,Mvec))
print(stardata)
```

```python
from sklearn.cluster import KMeans

clusterer = KMeans(n_clusters = 6)

clusterer.fit(stardata)
```

```python
plt.scatter(Tvec,Lvec, c = clusterer.labels_)
plt.xlabel("Temperature in Kelvin")
plt.ylabel("Luminosity of star")
plt.show()
```

```python
plt.scatter(Rvec,Mvec, c = clusterer.labels_)
plt.xlabel("Star radius")
plt.ylabel("Absolute magnitude")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scaler = StandardScaler()
```

```python
A_features = stars[stars.columns[0:4]]
print(A_features)
```

```python
scaled = scaler.fit_transform(A_features)
print(scaled)
```

```python
pca = PCA(n_components = 2)
pca.fit(scaled)
```

```python
S = pca.transform(scaled)
```
